```python
import pandas as pd
import matplotlib.pyplot as plt

# Creating sample dataset directly in code
data = {
    'sales': [200, 250, 300, 280, 320],
    'profit': [20, 30, 45, 40, 50]
}

df = pd.DataFrame(data)

# Quick data check
df.head()
```

```python
df.describe()
```

```python
plt.figure()
plt.plot(df['sales'], df['profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
plt.show()
```
