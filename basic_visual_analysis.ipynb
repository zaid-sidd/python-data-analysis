```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
    'sales': [200, 250, 300, 280, 320, 350],
    'profit': [20, 30, 45, 40, 50, 60]
}

df = pd.DataFrame(data)
df.head()
```

```python
df.describe()
```

```python
plt.figure()
plt.hist(df['sales'])
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure()
plt.scatter(df['sales'], df['profit'])
plt.xlabel('Sales')
plt.ylabel('Profit')
plt.title('Sales vs Profit')
plt.show()
```
